```python
import sqlite3
import pandas as pd
conn = sqlite3.connect("factbook.db")
```

```python
q = "SELECT * FROM sqlite_master WHERE type= 'table';"
pd.read_sql_query(q, conn)
```

```python
query = "select * from facts limit 5;"
pd.read_sql_query(query, conn)
```

```python
q_minpopulation = "select name, MIN(population) minpopulation from facts;"
pd.read_sql_query(q_minpopulation, conn)
```

```python
q_maxpopulation = "select name, MAX(population) maxpopulation from facts;"
pd.read_sql_query(q_maxpopulation, conn)
```

```python
q_minpopgrowth = "select name, MIN(population_growth) minpopgrowth from facts;"
pd.read_sql_query(q_minpopgrowth, conn)
```

```python
q_maxpopgrowth = "select name, MAX(population_growth) maxpopgrowth from facts;"
pd.read_sql_query(q_maxpopgrowth, conn)
```

```python
q_popzero = "select name from facts where population = 0;"
pd.read_sql_query(q_popzero, conn)
```

```python
q_popmax = "select name from facts where population = 7256490011;"
pd.read_sql_query(q_popmax, conn)
```

World and Antartica are listed in the table as countries with maximum and minimum population.


```python
q_final = "select population, population_growth, birth_rate, death_rate from facts where population != 0 and population != 7256490011;"
pd.read_sql_query(q_final, conn)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

pd.read_sql_query(q_final, conn).hist(ax=ax)
```

Which countries have the highest population density?

```python
q_highestdensity = "select name, cast(population as float)/cast(area as float) as density from facts order by density desc limit 20;"
pd.read_sql_query(q_highestdensity, conn)
```

Histogram of population densities

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
pd.read_sql_query(q_highestdensity, conn).hist(ax=ax)
```

highest ratios of water to land

```python
q_waterlandratio = "select name, cast(area_water as float)/cast(area_land as float) waterlandratio from facts order by waterlandratio desc limit 20;"
pd.read_sql_query(q_waterlandratio, conn)
```

```python
q_morewaterthanland = "select name from facts where area_water > area_land order by name desc;"
pd.read_sql_query(q_morewaterthanland, conn)
```
